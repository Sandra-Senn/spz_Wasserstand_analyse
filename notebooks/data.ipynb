## Wasserstand

https://zenodo.org/records/15025258

```python
df_location = pd.read_csv(r'..\data\wasserstand\Caravan_extension_CH\attributes\camelsch\attributes_other_camelsch.csv')
df_location
orte = df_location.loc[df_location['country'] == 'CH', ['gauge_name', 'gauge_id']]
orte
```

Die wichtigsten Kriterien bei der Auswahl sind:
- Keine Regulierung (keine Stauwerke, Dämme oder Begradigungen), um natürliche zeitliche Muster zu haben
- Langjährige, kontinuierliche Messreihen mit Wasserstand, Temperatur und Niederschlagsdaten.
- Regionale Varianz und typische Abflussregime (z. B. glazial, nival oder pluvial).

Aus der Liste scheinen einige bekannte natürliche Flüsse und Seenorte dabei zu sein. Gute Kandidaten sind zum Beispiel alpine oder voralpine Flüsse mit naturnahem Zustand, z.B. Orte wie:
- Brienzwiler (Alpenvorland)
- Thun (am Thunersee, der teilweise reguliert ist, aber Flusszuflüsse unreguliert sein können)
- Emmenmatt (Vorland)
- Ilanz (Rheingebiet, eher natürlich)
- Andermatt (Hochalpin, wahrscheinlich natürlich)
- Sarnen (am Sarnersee, eventuell mäßig reguliert)

```python
orte_namen = ["Brienzwiler", "Thun", "Emmenmatt", "Ilanz", "Andermatt", "Sarnen"]

auswahl_orte = orte[orte["gauge_name"].isin(orte_namen)]
auswahl_orte
```
